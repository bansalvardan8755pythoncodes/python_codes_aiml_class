```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
name=["Rahul","Priya","Amit","Neha","Arjun","Sneha","Rohit","Pooja","Vikram","Anjali"]
employee_id = ["EMP001", "EMP002", "EMP003", "EMP004", "EMP005", "EMP006", "EMP007", "EMP008", "EMP009", "EMP010"]
experience=[1.8,2.9,3.6,4.5,5.8,6.3,7.5,8.9,9.9,10.2]
salary=[10000,23000,36000,45000,54367,67553,71212,86454,98765,99008]
```

```python
data={"name":name,
      "employee_id":employee_id,
      "experience":experience,
      "salary":salary}
df=pd.DataFrame(data)
df
```

```python
df.to_csv("ai_16march.csv",index=False)
```

```python
df=pd.read_csv("ai_16march.csv")
df
```

```python
x=df["experience"]
y=df["salary"]
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,marker="+",color="red",label="experience vs salary")
plt.title("graph")
plt.xlabel("experience")
plt.ylabel("salary")
plt.legend()
plt.show()
```

```python
x=np.array(x).reshape(-1,1)
y=np.array(y).reshape(-1,1)
```

```python
X_test,X_train,y_test,y_train=train_test_split(x,y,test_size=0.3
                                               ,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
