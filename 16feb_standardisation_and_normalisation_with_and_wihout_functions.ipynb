```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,StandardScaler 
```

```python
num=50
roll_no=[i for i in range(1,num+1)]
name=[f"student_{i}" for i in range(1,num+1)]
height=np.random.randint(150,190,size=50)
weight=np.random.randint(50,90,size=50)
```

```python
data={
    "roll":roll_no,
    "name":name,
    "height":height,
    "weight":weight
}
```

```python
df=pd.DataFrame(data,index=[i for i in range(1,num+1)])
```

```python
print(df)
```

```python
df_normalized=df.copy()
cols_to_normalise=["height","weight"]
for col in cols_to_normalise:
    col_min=df[col].min()
    col_max=df[col].max()
    df_normalized[col] = (df[col] - col_min) / (col_max-col_min)
print(df_normalized.head(2))
```

```python
mm=MinMaxScaler()
df2=pd.DataFrame(mm.fit_transform(df[["height","weight"]]))
df2
```

```python
ss=StandardScaler()
df3=pd.DataFrame(ss.fit_transform(df[["height","weight"]]))
df3
```

```python
df_standarized=df.copy()
cols_to_standardise=["height","weight"]
for col in cols_to_standardise:
    col_mean=df[col].mean()
    col_std=df[col].std()
    df_standarized[col] = (df[col] - col_mean) / (col_std)
print(df_standarized.head(2))
```
