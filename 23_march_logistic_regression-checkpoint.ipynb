```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import warnings
warnings.filterwarnings("ignore")
```

```python
np.random.seed(42)
n = 500
age = np.random.randint(20, 80, n)
blood_pressure = np.random.randint(90, 180, n)
cholesterol = np.random.randint(150, 300, n)
smoking = np.random.randint(0, 2, n)  # 0 = No, 1 = Yes
risk_score = (
    0.03 * age +
    0.02 * blood_pressure +
    0.015 * cholesterol +
    0.5 * smoking
)
Disease_Risk = []
for i in risk_score:
    if i > 10:
        Disease_Risk.append(1)
    else:
        Disease_Risk.append(0)
df = pd.DataFrame({
    "Age": age,
    "Blood_Pressure": blood_pressure,
    "Cholesterol": cholesterol,
    "Smoking": smoking,
    "Disease_Risk": Disease_Risk
})
df.to_csv("disease_risk.csv", index=False)
```

```python
df = pd.read_csv("disease_risk.csv")
```

```python
X = df.drop("Disease_Risk", axis=1)
y = df["Disease_Risk"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
print("\nModel Coefficients:")
i = 0
for col in X.columns:
    print(col, ":", model.coef_[0][i])
    i += 1
print("\nIntercept:", model.intercept_[0])
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy:", accuracy)
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)
```
