```python
#inclusion of various libraries
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
```

```python
np.random.seed(42)
n = 300
data = {
    "Study_Hours": np.random.randint(1, 10, n),
    "Attendance": np.random.randint(50, 100, n),
    "Assignments": np.random.randint(1, 10, n),
    "Previous_Score": np.random.randint(40, 100, n),
    "Sleep_Hours": np.random.randint(4, 9, n)
}
df = pd.DataFrame(data)
df["Result"] = (
    (df["Study_Hours"] * 2 +
     df["Attendance"] * 0.3 +
     df["Assignments"] * 2 +
     df["Previous_Score"] * 0.5) > 100).astype(int)
df.to_csv("student_performance.csv", index=False)
```

```python
df=pd.read_csv("student_performance.csv")
print("dataset read successfully")
```

```python
X = df.drop('Result',axis=1)
y = df['Result']  
```

```python
splits = {'70:30': 0.3, '60:40': 0.4, '80:20': 0.2}
for split_name, test_size in splits.items():
    print(f"\nTrain-Test Split {split_name}")
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=test_size, random_state=42, stratify=y)
    k = 5
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted')
    rec = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    cm = confusion_matrix(y_test, y_pred)
    print(f"Accuracy: {acc}")
    print(f"Precision: {prec}")
    print(f"Recall: {rec}")
    print(f"F1 Score: {f1}")
    print("Confusion Matrix:")
    print(cm)
new_students = pd.DataFrame({
    "Study_Hours": [6, 3],
    "Attendance": [80, 60],
    "Assignments": [7, 4],
    "Previous_Score": [75, 50],
    "Sleep_Hours": [7, 5]
})

predictions = knn.predict(new_students)
print("\nPredictions for new students:", predictions)
```
