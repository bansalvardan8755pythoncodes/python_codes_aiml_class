```python
import numpy as np
import pandas as pd
roll_no=np.arange(1,11)
marks=np.array([78,85,90,66,72,88,91,69,75,80])
attendance=np.array([85,90,95,70,80,92,96,75,88,90])
gender=np.array(["m","f","m",None,"f","m",None,"f","m","f"])
student_df=pd.DataFrame({"roll no":roll_no,
                         "marks":marks,
                         "attendance":attendance,
                         "gender":gender})
print(student_df)
```

```python
null_count=student_df["gender"].isnull().sum()
print("null values in gender column:",null_count)
gender_mode=student_df["gender"].mode()[0]
student_df["gender"].fillna(gender_mode,inplace=True)
print(student_df)
```

```python
avg_marks=student_df["marks"].mean()
print("average marks:",avg_marks)
gender_count=student_df["gender"].value_counts()
print(gender_count)
```

```python
import matplotlib.pyplot as plt
gender_count=student_df["gender"].value_counts()
plt.bar(gender_count.index,gender_count.values)
plt.xlabel("gender")
plt.ylabel("number of students")
plt.title("gender distrubition of students")
plt.legend(["students"])
plt.show()
```

```python
plt.hist(student_df["marks"],bins=5)
plt.xlabel("marks")
plt.ylabel("number of students")
plt.legend(["marks"])
plt.show()
```

```python
plt.hist(student_df["attendance"],bins=5)
plt.xlabel("attendance percentage")
plt.ylabel("number of students")
plt.title("attendance distribution of students")
plt.legend(["attendance"])
plt.show()
```
