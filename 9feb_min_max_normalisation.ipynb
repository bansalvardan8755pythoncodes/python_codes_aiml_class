```python
import pandas as pd
import numpy as np
np.random.seed(42)
num_students = 50
data = {
    "Roll No": np.arange(1, num_students + 1),
    "Name": [f"Student_{i}" for i in range(1, num_students + 1)],
    "Height(cm)": np.random.randint(150, 190, size=num_students),
    "Weight(kg)": np.random.randint(50, 90, size=num_students),
    "Stipend": np.random.randint(5000, 15000, size=num_students)
}
df = pd.DataFrame(data)
print("Original Dataset (First 2 Rows)")
print(df.head(2))
print("\n")
df_normalized = df.copy()


cols_to_normalize = ["Height(cm)", "Weight(kg)", "Stipend"]

for col in cols_to_normalize:
    col_min = df[col].min()
    col_max = df[col].max()
    df_normalized[col] = (df[col] - col_min) / (col_max - col_min)

print("Normalized Data (First 2 Rows)")
print(df_normalized.head(2))
```
