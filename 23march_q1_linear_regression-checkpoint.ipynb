```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
np.random.seed(42) 
n = 100
years_of_experience = np.random.randint(1, 21, n)
salary = 5000 * years_of_experience + 30000 + np.random.normal(0, 5000, n)
df = pd.DataFrame({
    "years_of_experience": years_of_experience,
    "salary": salary
})
df.to_csv("salry_prediction.csv", index=False)
```

```python
df=pd.read_csv("salry_prediction.csv")
print("first two rows:")
print(df.head(2))
```

```python
X = df[["years_of_experience"]]
y = df["salary"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print("\nSlope (m):", slope)
print("Intercept (c):", intercept)
```

```python
plt.scatter(X, y, color="blue", label="Actual Data")
y_pred_line = model.predict(X)
plt.plot(X, y_pred_line, color="red", label="Regression Line")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.legend()
plt.show()
```

```python
new_exp = np.array([[2], [5], [10]])
predicted_salary = model.predict(new_exp)
print("\nPredicted Salaries:")
for i in range(len(new_exp)):
    print(new_exp[i][0], predicted_salary[i])

# Predictions on test data
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\nModel Evaluation:")
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
