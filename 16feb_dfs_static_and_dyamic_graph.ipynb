```python
def dfs(start,goal,graph,vis=None):
    if vis is None:
        vis=set()
    print(start,end=" ")
    if start==goal:
        return True
    vis.add(start)
    for child in graph[start]:
        if child not in vis:
            if dfs(child,goal,graph,vis):
                return True
    return False
```

```python
graph={
    "A":["B","C"],"B":["D","E","F"],"C":["D","F"],"D":["E","F"],"E":["F"],"F":[]
}
dfs("A","F",graph)
```
