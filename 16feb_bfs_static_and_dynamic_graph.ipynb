```python
def bfs(start,goal,graph):
    vis=set()
    path=[]
    queue=[start]
    vis.add(start)
    while queue:
        node=queue.pop(0)
        path.append(node)
        if node==goal:
            return path

        for child in graph[node]:
            if child not in vis:
                queue.append(child)
                vis.add(child)
    return False
```

```python
graph={
    "A":["B","C"],"B":["D","E","F"],"C":["D","F"],"D":["E","F"],"E":["F"],"F":[]
}
```

```python
bfs("A","F",graph)
```
